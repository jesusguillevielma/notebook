```python
import requests

from bs4 import BeautifulSoup

 

page = requests.get("http://www.nationmaster.com/country-info/stats/Media/Internet-users#")

page

 

soup = BeautifulSoup(page.content,'html.parser')

 

table = soup.findAll('table')

table = table[0]

 

for row in table.findAll('tr'):

    new_row=[]

    for cell in row.findAll('td'):

        new_row.append(cell.get_text().strip())

    print (new_row)

 

with open('nuevo.csv', 'a') as f:

    f.write('#; Country; Amount; Date;Graph; History\n')

    for row in table.findAll("tr"):

        new_row = []

       

        

        for cell in row("td"):

            new_row.append(cell.get_text().strip())

       

        # write list to csv file

        if len(new_row) >2 :

            f.write('{}; {}; {}; {}; {};{}\n'.format(new_row[0], new_row[1], new_row[2], new_row[3], new_row[4],new_row[5]))

            print('{}; {}; {}; {};{}; {}; \n'.format(new_row[0], new_row[1], new_row[2], new_row[3], new_row[4],new_row[5]))

 

```
