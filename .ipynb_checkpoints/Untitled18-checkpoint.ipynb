```python
import requests
from bs4 import BeautifulSoup 
page = requests.get('https://en.wikipedia.org/wiki/List_of_countries_by_wealth_per_adult')
page
```

```python
soup = BeautifulSoup(page.text,'html.parser')
table = soup.findAll('table')
```

```python
table = table[1]
table
for row in table.findAll('tr'):
    new_row = []
    
    for cell in row.findAll('td'):
        new_row.append(cell.get_text().strip())
    print(new_row)
```

```python
with open('rico.csv', 'a') as f:

    f.write('Coubtry;Wealth\n')

    for row in table.findAll("tr"):

        new_row = []

       

        

        for cell in row("td"):

            new_row.append(cell.get_text().strip())

       

        # write list to csv file

        if len(new_row) >0 :

            f.write('{}; {}\n'.format(new_row[0], new_row[1]))

            print('{}; {} \n'.format(new_row[0], new_row[1]))
```
