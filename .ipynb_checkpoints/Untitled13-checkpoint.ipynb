```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_postcode_districts_in_the_United_Kingdom')
page
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
table = soup.findAll('table')
```

```python
table = table[1]
table
for row in table.findAll('tr'):
    new_row= []
    for cell in row.findAll('td'):
        new_row.append(cell.get_text().strip())
    print(new_row)
```

```python
with open('population.csv', 'a') as f:

    f.write('Country;Population;Yearly;Net_Change;Density;Land_Area\n')

    for row in table.findAll("tr"):

        new_row = []

       

        

        for cell in row("td"):

            new_row.append(cell.get_text().strip())

       

        # write list to csv file

        if len(new_row) >2 :

            f.write('{}; {}; {}; {}; {};{}\n'.format(new_row[0], new_row[1], new_row[2], new_row[3], new_row[4],new_row[5]))

            print('{}; {}; {}; {};{}; {}; \n'.format(new_row[0], new_row[1], new_row[2], new_row[3], new_row[4],new_row[5]))
```
